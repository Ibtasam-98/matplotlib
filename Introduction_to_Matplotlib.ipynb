# Introduction to Matplotlib

Matplotlib is a Python library for creating static, interactive, and animated visualizations like plots and charts for data exploration, analysis, and presentation.

## What is Matplotlib and its uses?

Matplotlib is a Python library for creating visualizations.

1. Data exploration and analysis through plots
2. Generating publication-quality figures.
3. Visualizing various data types (line, bar, scatter, etc.).
4. Customizing plot appearance.
5. Integration with NumPy and Pandas.
6. Embedding plots in applications.
7. Creating interactive plots and animations.









```python
import matplotlib.pyplot as plt
import pandas as pd
```

# Basic Plotting

```python
# Sample data
x = [1,2,3,4,5]
y = [6,7,8,9,10]

# Creating basic plot
plt.plot(x,y)
plt.xlabel("X-Axix")
plt.ylabel("Y-Axis")
plt.title("Basic Plot")

plt.show()

```

## Saving the plot to files

```python
plt.savefig("line_plot.png")
plt.savefig("line_plot.jpg")
plt.savefig("line_plot.jpeg")
plt.savefig("line_plot.svg")
plt.savefig("line_plot.pdf")

print("Plots saved to Colab's file storage.")
```

# Scatter Plot

```python
# Sample data
x = [1,2,3,4,5]
y = [6,7,8,9,10]


# Creating scatter plot
plt.scatter(x,y)
plt.xlabel("X-Axix")
plt.ylabel("Y-Axis")
plt.title("Basic Scatter Plot")
plt.show()
```

# Bar Chart (Verticle)

```python
# Sample data
x = [1,2,3,4,5]
y = [6,7,8,9,10]


# Creating Vertical Bar Chart
plt.bar(x,y)
plt.xlabel("X-Axix")
plt.ylabel("Y-Axis")
plt.title("Vertical Bar Chart")
plt.show()
```

# Bar Chart (Horizontal)

```python
# Sample data
x = [1,2,3,4,5]
y = [6,7,8,9,10]


# Creating Horizontal Bar Chart
plt.barh(x,y)
plt.xlabel("X-Axix")
plt.ylabel("Y-Axis")
plt.title("Horizontal Bar Chart")
plt.show()
```

# Histogram

```python

# Sample data
x = [1,2,3,4,5]
y = [6,7,8,9,10]

plt.figure(figsize=(8, 6))
plt.hist(x, bins=len(set(y)), edgecolor='black')
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Histogram of y_data")
plt.show()
```

# Pie Chart

```python
x = [1, 2, 3, 4, 5]
y = [6, 7, 8, 9, 10]

# Pie for x
plt.figure(figsize=(4, 4))
plt.pie(x, labels=x, autopct='%1.1f%%', startangle=90)
plt.title("Pie of x")
plt.axis('equal')
plt.show()

# Pie for y
plt.figure(figsize=(4, 4))
plt.pie(y, labels=y, autopct='%1.1f%%', startangle=90)
plt.title("Pie of y")
plt.axis('equal')
plt.show()
```
